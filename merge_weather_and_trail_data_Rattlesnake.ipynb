## Load DataSets

```python
import pickle
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Import weather data - Cedar Lake - shorter version
with open('cedar_lake_weather_from_2010.pickle','rb') as read_file:
   cedar_weather_short = pickle.load(read_file)

#now, convert the weather data to a dataframe, add column headers, and convert date string to datetime
cedar_weather_short = pd.DataFrame(cedar_weather_short,columns=['timestamp','Ced_Temp-Max','Ced_Temp-Min','Ced_Temp-Avg','Ced_Departure','Ced_HDD','Ced_CDD'
                                              ,'Ced_Precipitation','Ced_New_Snow','Ced_Snow_Depth'])

cedar_weather_short.head()
```

```python
# Import weather data - Seattle 
with open('seattle_weather_2010_to_2019.pickle','rb') as read_file:
   seattle_weather = pickle.load(read_file)

#now, convert the weather data to a dataframe, add column headers, and convert date string to datetime
seattle_weather = pd.DataFrame(seattle_weather,columns=['timestamp','Sea_Temp-Max','Sea_Temp-Min','Sea_Temp-Avg','Sea_Departure','Sea_HDD','Sea_CDD'
                                              ,'Sea_Precipitation','Sea_New_Snow','Sea_Snow_Depth'])

seattle_weather.head()
```

```python
seattle_weather.info()
```

```python
# Import trail data
with open('rattlesnake.pickle', 'rb') as read_file:
    df = pickle.load(read_file)
    
df.head()
```

## Merge Weather DataSets into Trails DataSet

### Convert timestamp on trails into weather data format. 
2015-01-01
Year - Month - Day (string)

```python
df['year'] = df['report_date'].apply(lambda report_date: report_date[-4:])
```

```python
df['day'] = df['report_date'].apply(lambda report_date: report_date[-8:-6])
```

```python
df['month'] = df['report_date'].apply(lambda report_date: report_date[:3])

months_dict = {'Jan':'01', 'Feb':'02', 'Mar':'03', 
               'Apr':'04', 'May':'05', 'Jun':'06',
               'Jul':'07', 'Aug':'08', 'Sep':'09',
               'Oct':'10', 'Nov':'11', 'Dec':'12'}

df['month']= df['month'].map(months_dict)
```

```python
df['timestamp'] = df['year'] + '-' + df['month'] + '-' + df['day']
df.head()
```

```python
df['year'].value_counts()
```

```python
df2 = pd.merge(df,seattle_weather,on='timestamp')
df2.head()
```

```python
df3 = pd.merge(df2,cedar_weather_short,on='timestamp')
df3.head()
```

```python
with open('rattlesnake_trails_and_weather.pickle', 'wb') as to_write:
    pickle.dump(df3, to_write)
```
