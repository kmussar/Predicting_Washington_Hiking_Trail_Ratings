## Remove Unneccessary Columns: 
* HDD (Sea and Cedar)
* CDD (Sea and Cedar)
* num_reports
* url


HDD (Heating Degree Day) is the number of degrees that a day's average temperature is below 65 Fahrenheit (18 Celsius), which is the temperature below which buildings need to be heated. 

CDD (Cooling Degree Day) is a measurement designed to quantify the demand for energy needed to cool a building. It is the number of degrees that a day's average temperature is above 65 Fahrenheit (18 Celsius), which is the temperature above which buildings need to be cooled. 
-- https://www.investopedia.com/terms/c/colddegreeday.asp

```python
df.drop(columns=['num_reports','url','Sea_HDD','Sea_CDD', 'Ced_HDD', 'Ced_CDD','report_date'],axis=1,inplace=True)
df.head()
```

### Convert string dates to timestamp object 

```python
df['datetime'] = pd.to_datetime(df['timestamp'])
df.head()
```

### Add new column - day of week (name) 

```python
df['day_of_week'] = df['datetime'].dt.day_name()
df.head()
```

### Convert columns to floats
* rating (also index first number)
* year
* day 
* month
* Sea_temp_Max
* Sea Temp Min 
* ... Ced Snow Depth

Note: Some values include T and M instead of numbers. T means trace and M means missing data. I will replace T with 0 and M with NaN. 

```python
def replace_T_M(string): 
    string.replace("T  ",'0')
    string.replace("T",'0')
    string.replace("M",'')
    string.replace("M  ",'')
    return string
```

```python
df['Ced_Precipitation'] = df['Ced_Precipitation'].apply(replace_T_M)
df['Ced_Precipitation'].unique()
```

```python
# SWITCHED TO THIS APPROACH 

# Convert several columns to floats: 
columns_to_convert = ['Ced_Temp-Max','Ced_Temp-Min', 'Ced_Temp-Avg', 'Ced_Departure','Ced_Precipitation','Ced_New_Snow', 'Ced_Snow_Depth']
#'Sea_Precipitation','Sea_New_Snow', 'Sea_Snow_Depth', 

for i in range(len(columns_to_convert)):
    df[columns_to_convert[i]] = df[columns_to_convert[i]].str.replace("T",'0')
    df[columns_to_convert[i]] = df[columns_to_convert[i]].str.replace("M",'')
    df[columns_to_convert[i]] = pd.to_numeric(df[columns_to_convert[i]])
df.info()
```

```python
df['rating'] = df['rating'].apply(lambda rating: rating[:3])
df['rating'] = pd.to_numeric(df['rating'])
```
